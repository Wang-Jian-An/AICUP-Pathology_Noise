# 卷積神經網路(Convolutional Neural Network) 基本觀念說明

- 何謂卷積運算 (Convolution)？
    - 透過一具代表性的濾波器 (Filter; Kernel)，將特定區塊的圖像範圍彙整成一具代表性的數值。
    - 依據由圖像的左上角開始，由左到右、由上到下，計算出一張具代表性特徵的圖像，俗稱 Feature map。
    - 卷積運算後，圖像大小的變化→一定會 **縮小**
    - 如何達成卷積運算後維持與原始圖片大小相同？→使用 **zero-padding**
